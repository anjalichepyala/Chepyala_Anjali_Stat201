```python
string = "abcabcabc"
substrings = set()

for i in range(len(string) - 2):
    littlestring = string[i:i+3]
    substrings.add(littlestring)

print(f"Number of distinct 3-letter substrings: {len(substrings)}")
```

```python
string1 = "racecaranna"
count = 0
found = set()
for length in range(3, len(string1)+1, 2):
    for i in range(len(string1) - length + 1):
        substring1 = string1[i:i+length]
        if substring1 == substring1[::-1]:
            if substring1 not in found:
                found.add(substring1)
                count += 1

print(count)
```

```python
number1 = input("First number:")
number2 = input("Second number:")
start = int(number1)
end = int(number2)

for num in range(start, end +1):
    if num>1:
        is_prime = True
        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            print(num)
```
