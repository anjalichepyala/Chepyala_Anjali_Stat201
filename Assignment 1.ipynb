```python
#Question 3
while True:
    num_emails = input ("How many email addresses do you want to input")
    try:
        num_emails == float(num_emails)
        break
    except ValueError:
        print("Invalid input! Please enter a numerical value")

email_num = int(num_emails)
professor_email = ""
student_email = ""
for i in range(email_num):
    email = input("Enter an email address:").lower().strip()
    if email.endswith("@u.northwestern.edu"):
        student_email += email + "\n"
    elif email.endswith("@northwestern.edu"):
        professor_email += email + "\n"
        
print("\nProfessor Emails:")
if professor_email == "":
    print("None")
else:
    print(professor_email.strip())

print("\nStudent Emails:")
if student_email == "":
    print("None")
else:
    print(student_email.strip())

```
