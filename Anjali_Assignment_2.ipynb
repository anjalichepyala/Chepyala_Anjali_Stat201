b) Write a program where you play "Rock, Paper, Scissors" games against the computer until you have won 3 games. Simulate a random choice for the computer each game and ask the user to enter either rock, paper, or scissors. Print the number of games it took to win 3 times. Show the output of the program after playing the game.

Example:

                                    "You won 3 out of <num_games> games."

```python
import random as rm 

user_wins = []       #makes an empty list for user_wins
game_count = 1       #keeps track of how many games have been played

#loops until user wins 3 games
while len(user_wins) < 3:
    #computer randomly chooses rock, paper, or scissors
    comp_choice = rm.choice(['rock', 'paper', 'scissors'])

    #asks user for input
    while True:
        user_input = input(f"Game {game_count} - Rock, Paper, Scissors, shoot! ").lower().strip()
        if user_input in ["rock", "paper", "scissors"]:
            break  #prints invalid input if user inputs somethind other than rock or paper or scissors
        else:
            print("Invalid input! Please enter rock, paper, or scissors.")

    #prints results of choices in game 3
    print(f"In game {game_count}, the computer picked {comp_choice}!")
    print(f"In game {game_count}, you picked {user_input}!")

    #adds 1 to the list if user won round
    if user_input == "rock" and comp_choice == "scissors":
        user_wins.append(1)
    elif user_input == "paper" and comp_choice == "rock":
        user_wins.append(1)
    elif user_input == "scissors" and comp_choice == "paper":
        user_wins.append(1)

    #moves to the next game
    game_count += 1

#prints final result after user wins 3 times
print(f"You won 3 games out of {game_count - 1} games")
```

### Question 5 (6 points)

a) Write a program (loop) that identifies whether a positive integer input by the user is prime or not. A prime number is a number whose only divisors are 1 and itself. Show the output when the program is used to check if `89` is prime or not.

```python
while True:
    number_input = input("Input a number to check if it's prime: ").strip()
    #asks for user input for a number and strips it
    try:
        number_input = int(number_input) 
        break  #exits loop if input is valid or prints invalid input if not
    except ValueError:
        print("Invalid input! Please enter a numerical whole value.")

if number_input < 2:
    is_prime = False
    #automatically sorts for numbers less than 2
else:
    is_prime = True  #checks to see if number is prime by checking if number has factors
    for i in range(2, number_input):
        if number_input % i == 0:
            is_prime = False
            break #breaks once loop is completed after loop reaches inputted number

if is_prime:
    #prints statement if number is prime or not
    print(f"{number_input} is a prime number.")
else:
    print(f"{number_input} is not a prime number.")
```

### Question 6 (4 points)

Write a function that calculates the area of a rectangle. The function should have an input for length (in inches) and width (in inches). And have 2 ouputs: the area in terms of square inches and the area in terms of square feet. There are 12 inches in 1 foot. Run your function with a length of 102 inches and width of 60 inches to check if it works.

```python
#converts inputed length in inches to a float, strips, and sets to length
length = float(input("What is the length of the rectangle (in)?").strip())
#converts inputed width in inches to a float, strips, and sets to length
width = float(input("What is the width of the rectangle (in)?").strip())
#changes length in feet to inches
ft_length = float(length)/12
#changes width in feet to inches
ft_width = float(width)/12
#calculates area in ft
area_ft = ft_length * ft_width
#calculates area in inches
area_in = length * width
#prints final statement
print(f"The area of the rectangle in square ft is {area_ft} square feet and the area of the rectangle in square inches is {area_in} square inches")
```
